```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import expon # Exponential distribution
```

# Distribución Exponencial

**Función de Densidad**
$$ f(x) = \lambda e^{-\lambda x} $$

Donde:
- $x \geq 0$
- $\lambda > 0$

**Función de Distribución**
$$ F(x) = 1 - e^{-\lambda x} $$

# Ejercicio

La duración de una llamada telefónica es una variable aleatoria con distribución exponencial de media $\lambda = 3$ minutos. ¿Cuál es la probabilidad de que una llamada dure menos de 2 minutos?

## Graficamos usando la formula metematica

```python
lambda_ = 1/3

f = lambda x: lambda_ * np.exp(-lambda_ * x)

# Graficamos
x = np.linspace(0, 10, 1000)
y = f(x)

plt.plot(x, y, label='lambda = 3')
plt.legend()
plt.grid()
plt.show()
```

## Calculamos la probabilidad de que una llamada dure menos de 2 minutos

```python
# Parámetro lambda de la distribución exponencial (media)
lambd = 1 / 3  # Media de 3 minutos

# Crear objeto de la distribución exponencial
distribucion_exponencial = expon(scale=1/lambd)

# Calcular la probabilidad de que una llamada dure menos de 2 minutos
probabilidad_menos_2_minutos = distribucion_exponencial.cdf(2)

print("Probabilidad de que una llamada dure menos de 2 minutos:", probabilidad_menos_2_minutos)
```

## Simulamos muestras aleatorias de la distribución exponencial

```python
cantidad_simulaciones = 100
```

```python
# Generar una muestra aleatoria de duración de llamada
muestra_llamada = distribucion_exponencial.rvs(size=cantidad_simulaciones)
```

Usamos la regla de [Regla de Sturges](https://es.wikipedia.org/wiki/Regla_de_Sturges) para determinar el número de intervalos en un histograma.

$$ c = 1 + \log_{2}{M}$$

$$ c = 1 + \log_{10}{M} / \log_{10}{2}$$

Donde:
- $c$ es el número de intervalos
- $M$ es el número de muestras

```python
# grafica mos la muestra_llamada
plt.hist(muestra_llamada, bins=7, density=True, histtype='step', color='blue', label='Distribución de llamadas')

# Graficar la función de densidad de probabilidad
x = np.linspace(0, 10, 1000)
y = distribucion_exponencial.pdf(x)

plt.plot(x, y, 'r--', label='PDF')
plt.grid()
plt.show()
```
