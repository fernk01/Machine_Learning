```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Crear un DataFrame de ejemplo
data = {
    'Pressure9am': [1012, 1015, np.nan, 1013, 1014],
    'Pressure3pm': [1010, 1013, 1012, np.nan, 1015]
}
df = pd.DataFrame(data)

# Definir el IterativeImputer
imputer = IterativeImputer()

# Aplicar el imputador
imputed_data = imputer.fit_transform(df)

# Convertir el resultado a un DataFrame para visualizarlo mejor
imputed_df = pd.DataFrame(imputed_data, columns=['Pressure9am', 'Pressure3pm'])

print(imputed_df)
```

```python
# graficomos los datos originales y los imputados
import matplotlib.pyplot as plt

plt.scatter(imputed_df['Pressure9am'], imputed_df['Pressure3pm'], color='blue', label='Imputado')
plt.scatter(df['Pressure9am'], df['Pressure3pm'], color='red', label='Original')

plt.xlabel('Pressure9am')
plt.ylabel('Pressure3pm')
plt.legend()
plt.show()
```
