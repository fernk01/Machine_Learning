# heatmaps 

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib
import matplotlib as mpl


vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(harvest)

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(farmers)), labels=farmers)
ax.set_yticks(np.arange(len(vegetables)), labels=vegetables)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```

Supongamos que la variable aleatoria `X` se distribuye simétricamente alrededor de cero (por ejemplo, una distribución normal estándar), y `Y = X^2`. Aunque `X` y `Y` están perfectamente dependientes, su correlación es cero. Aquí está el código:

Cuando ejecutes este código, verás que la correlación entre `X` y `Y` es cercana a cero, a pesar de que `Y` está completamente determinada por `X`. Esto se debe a que la correlación **mide la relación lineal** (si graficamos 2d se acerca una linea) entre `X` y `Y`, y la relación entre `X` y `Y` en este caso no es lineal. Espero que esto te ayude! Si tienes más preguntas, no dudes en preguntar.

```python
# Generamos 1000 muestras de la variable aleatoria X
X = np.arange(0, 10)
# X = np.random.normal(0, 1, 1000)  # con Y = X**2 da casi 0
# X = np.random.laplace(loc=0, scale=1, size=1000) # con Y = X**2 da casi 0

# Y es completamente determinada por X
Y = X**2
#Y = np.sin(X)

# Calculamos la correlación entre X y Y
correlacion = np.corrcoef(X, Y)[0,1]

print(f"La correlación entre X y Y es: {correlacion}")

# Graficamos X y Y
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def jitter_dots(dots):
    offsets = dots.get_offsets()
    jittered_offsets = offsets
    jittered_offsets[:, 0] += np.random.uniform(-0.3, 0.3, offsets.shape[0])
    dots.set_offsets(jittered_offsets)

df_1 = pd.DataFrame({
    'Cells': np.arange(0, 100),
    'Delta_7': np.random.rand(100),
    'Delta_10': np.random.rand(100),
    'Delta_14': np.random.rand(100)
})

fig, ax1 = plt.subplots()

columns = df_1.columns[1:]
c = df_1['Delta_7']
for i, column in enumerate(columns):
    dots = plt.scatter(np.full((len(df_1), 1), i), df_1[column], s=50, c=c, cmap='plasma')
    jitter_dots(dots)

plt.xticks(range(len(columns)), columns)
xmin, xmax = plt.xlim()
plt.xlim(xmin - 0.3, xmax + 0.3)  # make some room to show the jittered dots
cbar = plt.colorbar()
plt.show()
```
